#### **Statistical Difference**

```python
red_wine_df.columns
```

```python
features =['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']

for feature in features:
    red_wine_mean = red_wine_df[feature].mean()
    white_wine_mean = white_wine_df[feature].mean()
    diff = red_wine_mean  - white_wine_mean
    print(f"========================== {feature} ========================== ")
    print(f"Red wine has a avg of :{red_wine_mean:.3f} ")
    print(f"White wine has a avg of :{white_wine_mean:.3f} ")
    print(f"Diffence between Red & white wine averages is :{diff:.3f} ")
    print(f"Red wine is {'higher' if diff >0 else 'lower'} than white wine in terms of **{feature.upper()}** LEVELS")
    print(f"=============================================================== \n\n")
```

```python
reduced_white_wine_df = white_wine_df[:1599]
```

```python
#  diagram size
fig, axes = plt.subplots(2,3, figsize=(15,10))

#  target features
features =['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']

for i, feature in enumerate(features):
    row, col = i //2, i %2 

    # Red on left side
    axes[row, col].hist(red_wine_df[feature], bins=30, alpha=0.7, color="red", density=True) 
    # white woine on the right had side
    axes[row, col].hist(reduced_white_wine_df[feature], bins=30, alpha=0.7, color="green", density=True) 
    axes[row, col].set_title(feature)
    axes[row, col].legend()

plt.tight_layout()
plt.show()
```
