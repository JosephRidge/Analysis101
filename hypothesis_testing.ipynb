# Steps in conducting a hypothesis test: 
- Formulate the Null and Alternative Hypothesis
- Gather evidence(data)
- Calculate the test statistic
- Make a conclusion

## **Using Randomly Generated Data:**

**Problem:**
A Telecom service provider claims that individual customers pay on an average 400 bob per month with standard deviation of 25 bob A random sample of 50 customers bills during a given month is taken with a mean of 250 and standard deviation of 15.

summary:A telecom company claims that customer pay 400 bob a month on average for internet. We take 50 random bills and find the average is 250 bob permonth. Is the company lying?

- Claim: avg = 400
- Based on the sample: 
     - n =50
     - sample mean = 250 
     - sample sd = 15 



**Step 1: Formulate the Hypotheses:**
- Null (claim): mean = 400
- Alternative: mean != 400

- note: we are dealing with a *two tailed test* and it is numerical in nature hence the ideal test would be a t-test 

#### **Import Necessary Libraries**

```python
import numpy as np 
import pandas as pd
from scipy import stats
from sklearn.datasets import load_iris
```

#### **Generate/ Simulate the data(evidence/ data)**

```python
np.random.seed(42) # it generates pseudo random numbers = > Mathematical formular (consitency or rather enable reproducability), 42 states that i willbe statring from the 42nd position
bills = np.random.normal(250, 15, 50) # mean -250, sd - 15, n - 50
```

```python
output =f"We have generated {len(bills)} random bills"
print(output)
```

#### **Sample Statistic:**

```python
mean = bills.mean()
sd = bills.std()
n = len(bills)

output = f"MEAN: {mean:.3f} || Standard Deviation: {sd:.3f} || Sample Size: {n}"
print(output)
```

```python
t_stat, p_value = stats.ttest_1samp(bills, 400) # 400 is the claimed parameter value/ mean

output = f"Test Statistic: {t_stat:.3f} || p-value: {p_value:.3f}"
print(output)
```

#### **Conclusion:**

*Decision Rule:*
- if p< alpha => Reject the Null Hyp. (statistically significant)
- if p>= alpha => fail to reject the Null Hyp. (not statistically significant)

```python
# assume the level of significance(alpha) = 0.05
alpha = 0.05

output =""
if p_value < 0.05:
    output = "Reject the Null Hypothesis"
else:
    output = "Fail to reject the Null Hypothesis"

print(output)
```

#### **Load Dataset**

#### **Hypothesis:**

- Null(Ho): setosa petal length == versicolor petal length
- alternative(Ha): setosa petal length != versicolor petal length

- parameter of interest is *mean*
- the groups are independent
- two tailed test
- ideal test statistic here is t-test


#### **Seperate the groups:**

```python
setosa = df[df['species'] == 'setosa']['petal length (cm)']
versicolor = df[df['species'] == 'versicolor']['petal length (cm)']
virginica = df[df['species'] == 'virginica']['petal length (cm)']
```

#### **Calculate the mean | standard deviation | size**

```python
# length/ size of data sets
len_setosa = len(setosa) 
len_versicolor = len(versicolor)
len_virginica = len(virginica)


# mean
mean_setosa = setosa.mean()
mean_versicolor = versicolor.mean()
mean_virginica = virginica.mean()

# standard deviations
sd_setosa = setosa.std()
sd_versicolor = versicolor.std()
sd_virginica = virginica.std()

output_len = f"LEN** setosa: {len_setosa} |  versicolor: {len_versicolor}  | virginica: {len_virginica}"
output_mean = f"MEAN** setosa: {mean_setosa:.3f} |  versicolor: {mean_versicolor:.3f}  | virginica: {mean_virginica:.3f}"
output_sd = f"SD** setosa: {sd_setosa:.3f} |  versicolor: {sd_versicolor:.3f}  | virginica: {sd_virginica:.3f}"


print("=================================================")
print(output_len)
print(output_mean)
print(output_sd)

print("=================================================")
```

```python
setosa.describe()
```

```python
versicolor.describe()
```

```python
virginica.describe()
```

#### **Calculate the Test Statistic**

```python
t_stat, p_value = stats.ttest_ind(setosa, versicolor)
```

```python
output = f" Test Statistic {t_stat:.4f} | p-value: {p_value:.4f}"
print(output)
```

#### **Conclusion/ Statistical Significance**

```python
α = 0.05 # alpha

if p_value < α: 
    print("Reject Ho")
else:
    print("fail to reject Ho")
```

```python
t_stat, p_value = stats.ttest_ind(versicolor, virginica)
output = f" Test Statistic {t_stat:.4f} | p-value: {p_value:.4f}"
print(output)

α = 0.05 # alpha

if p_value < α: 
    print("Reject Ho")
else:
    print("fail to reject Ho")
```

#### **Hypothesis:**

- Null(Ho): the alcohol levels in red == white
- Alternative(Ha): the alcohol levels in red != white

parameter of interest is the mean

#### **Params:**

```python
def targetParams(df1, df2, targetColumn):
    # mean
    df1_mean = df1[targetColumn].mean()
    df2_mean = df2[targetColumn].mean()

    # standard deviation
    df1_sd = df1[targetColumn].std()
    df2_sd = df2[targetColumn].std()

    # length
    df1_len = len(df1[targetColumn]) 
    df2_len = len(df2[targetColumn]) 

    output_mean = f"MEAN** Red: {df1_mean:.3f} || White: {df2_mean:.3f}"
    output_sd = f"SD** Red: {df1_sd:.3f} || White: {df2_sd:.3f}"
    output_len = f"LEN** Red: {df1_len} || White: {df2_len}"

    print("==========================================")
    print(output_mean)
    print(output_sd)
    print(output_len)
    print("==========================================")
```

```python
targetParams(red_wine_df,white_wine_df,'alcohol')
```

#### **Calculate the Test Statistic**

```python
len(red_wine_df.columns)
```

```python
def getTstatisticAndConclude(df1, df2, targetColumn):
    t_stat, p_value = stats.ttest_ind(df1[targetColumn], df2[targetColumn])
    output = f"Test Statistic: {t_stat:.4f} | p-value: {p_value:.4f}"
    print(output)
    alpha = 0.05 

    if p_value < alpha:
        print("Reject Ho")

    else:
        print("Fail to reject Ho")

```

```python
getTstatisticAndConclude(red_wine_df, white_wine_df,'alcohol')
```
