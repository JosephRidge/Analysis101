### Two Events handling Probability Rules (AND and OR)

```python
number_of_experiements = 1000
die_one = np.random.randint(1,7, number_of_experiements)
die_two = np.random.randint(1,7, number_of_experiements)
```

```python
output = f"TNo | Die1| Die2 | Die1 equals 6? | Die2 equals 6?  | AND operation | OR operation "
print(output) 
for i in range(20):
    and_condition = (die_one[i] == 6) and (die_two[i] == 6) # STRICT contion such that die1 gives us 6 and die2 gives 6 as 
    or_condition = (die_one[i] == 6) or (die_two[i] == 6) # STRICT contion such that die1 gives us 6 and die2 gives 6 as 
    output = f"{i+1} | {die_one[i]:4d} | {die_two[i]:4d} | {die_one[i]==6} | {die_two[i]==6} | {and_condition} | {or_condition} "
    print(output)
```

#### **Calculate the probabilities**

Addition Rule:
P(A or B) = P(A) + P(B) - P(AnB)
p(A or B) = P(A) + P(B) => mutulaly exclusive events

Mutplication rule:
P(A and B) = P(A)P(B) => applies for independent events  
P(A and B) =P(A)P(A|B)


```python
prob_both_six = sum((die_one == 6) & (die_two == 6 ))/ number_of_experiements # & is a bitwise logical operators that enable us to use the Logical AND
prob_atleast_one_six = sum((die_one == 6) | (die_two == 6 ))/ number_of_experiements # & is a bitwise logical operators that enable us to use the Logical OR
```

```python
# results: 
print(f"Number of Rolls: {number_of_experiements}")
print(f"probability of AND scenario: {prob_both_six}")
print(f"Theoretical probability of AND scenario: 1/36")
print(f"probability of OR scenario: {prob_atleast_one_six}")
print(f"Theoretical probability of OR scenario: 11/36")
```

```python
plt.figure(figsize=(10, 8)) # manipulate the figue size based on length and width

probabilities = [prob_both_six, 1/36, prob_atleast_one_six, 11/36 ] # holds the probaility outputs (empirical  and theoretica;)
labels = ['AND\n(actual)', 'AND\n(theoretical)', 'OR\n(actual)' , 'OR\n(theoretical)'] # labels for the data
# colors =['blue', 'lightblue',' orange', 'green'] 

plt.bar(labels, probabilities, color=['lightblue', 'blue', 'green', 'orange'])

plt.title("AND vs OR ", fontsize=16)
plt.ylabel("Probability")
plt.ylim(0, 0.35)

for i, v in enumerate(probabilities):
    plt.text(i,v + 0.01,f"{v:.3f}", ha='center', fontsize = 12)

plt.show()
```
