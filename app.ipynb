##### Check points: 
- Displaying the first 5 rows of the dataset

Note: Kindly provide insights based on the above at the end of each data set.

```python
class DescriptiveStats:
    def __init__(self, dataset):
        self.dataset = dataset

    # get first 10:
    def dataHead(self, numOfRows):
        print(self.dataset.head(numOfRows))

    # last 10
    def dataTail(self, numOfRows):
        self.dataset.tail(numOfRows)

    # shape
    def dataShape(self):
        rows, columns = self.dataset.shape
        print(f"ROWS: {rows} || COLUMNS: {columns}")

    def dataTypes(self):
        print(self.dataset.dtypes)
    
    def descriptiveNumeical(self):
        print(self.dataset.describe())

    def descriptiveCategorical(self):
        self.dataset.describe(include='object')

    def getAllColumns(self):
        self.dataset.columns

```

```python
ratingsDescriptiveStats = DescriptiveStats(ratings_df)
```

```python
ratingsDescriptiveStats.dataHead(10)
```

```python
ratingsDescriptiveStats.descriptiveNumeical()
```

```python
ratingsDescriptiveStats.dataTypes()
```
