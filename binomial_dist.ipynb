```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot  as plt
from scipy import stats
import seaborn as sns
```

#### **Load Data**

```python
penguine_df = pd.read_csv("data/penguins_size.csv")
```

#### **Clean null values**

```python
penguine_df.isnull().sum()
```

```python
penguine_df = penguine_df.dropna()
```

```python
penguine_df.isnull().sum()
```

```python
penguine_df['sex'].unique()
```

```python
penguine_df = penguine_df[penguine_df['sex'] != '.']
```

```python
penguine_df['sex'].unique()
```

```python
penguine_df.shape
```

```python
male_count = (penguine_df['sex'] == 'MALE').sum()
print(male_count)
```

```python
# probaililty of attaining Male penguines
prob_male = male_count / len(penguine_df)
print(prob_male)
```

```python
n =20 # number of tials
p = prob_male # probability

x = np.arange(0, n+1) # values on x- axis
pmf = stats.binom.pmf(x, n , p)

plt.bar(x, pmf)
plt.title(f"Binomial Distribution (n ={n}, p={p})")
plt.xlabel("Number of successes(MALE)")
plt.ylabel("probabilities")
plt.show()
```

```python
female_count = (penguine_df['sex'] == 'FEMALE').sum()
print(female_count)# probaililty of attaining Male penguines
prob_female = female_count / len(penguine_df)
print(prob_female)
```

```python
n =20 # number of tials
p = prob_female # probability

x = np.arange(0, n+1) # values on x- axis
pmf = stats.binom.pmf(x, n , p)

plt.bar(x, pmf)
plt.title(f"Binomial Distribution (n ={n}, p={p})")
plt.xlabel("Number of successes(MALE)")
plt.ylabel("probabilities")
plt.show()
```
